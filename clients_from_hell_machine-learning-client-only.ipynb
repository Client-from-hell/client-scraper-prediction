```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = df[df['who'].str.contains('Client' )]
```

```python
df = df.replace(0, np.nan).dropna(axis = 1, how = 'all')
```

```python
df = df.fillna(0)
```

```python
df.head()
```

```python
df_ml = df.drop(columns = ['who', 'client_type'])
```

```python
df_ml.shape
```

```python
df_ml.head()
```

```python
df_ml_features = df_ml.drop(columns = 'Deadbeats')
df_ml_target = df_ml['Deadbeats']
```

```python
df_ml_features.head()
```

```python
def tf(bag_of_words):
    totals = bag_of_words.sum()
    n_words = totals.sum()
    return totals/n_words
tf(df_ml_features)
```

```python
from math import log

def idf(bag_of_words):
    n_docs = bag_of_words.shape[0]
    doc_freq = bag_of_words.astype('int').sum()
    return (n_docs / doc_freq).apply(log)
    
idf(df_ml_features)
```

```python
def tf_idf(bag_of_words):
    return tf(bag_of_words) * idf(bag_of_words)

tf_idf = tf_idf(df_ml_features).sort_values(ascending = False)
tf_idf
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier

regressors = {'Logistic Regression': LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr'),
              'Gaussian NB': GaussianNB(),
              'Bernoulli NB': BernoulliNB(),
              'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0),
              'KNeighborsClassifier': KNeighborsClassifier(), 
              'SVM': svm.LinearSVC(),
              'Neural Network':  MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)}

def compare_models(x, y, model_dict, folds=3):
    results = []
    for name, model in model_dict.items():
        scores = cross_val_score(model, x, y, cv=folds)
        stats = [name, scores.mean(), min(scores), max(scores), scores.std(), pd.Series(scores).mad()]
        results.append(stats)
    df = pd.DataFrame(results, columns = ['Model', 'Mean', 'Min', 'Max','Std', 'Mad'])
    df = df.sort_values('Mean', ascending = False)
    return df
```

```python
y = df_ml_target
x = df_ml_features
```

```python
compare_models(x, y, regressors, 5)
```

```python
q50 = tf_idf.quantile(0.50)
columns_remove = list(tf_idf[(tf_idf < q50)].index)
len(columns_remove)
```

```python
df_q50 = df_ml.drop(columns = columns_remove)
```

```python
df_q50.shape
```

```python
x50 = df_q50.drop(columns = 'Deadbeats')
y50 = y
```

```python
compare_models(x50, y50, regressors, 5)
```

```python
df_2000 = df_ml[list(tf_idf[:2000].index)]
```

```python
x_2000 = df_ml.drop(columns = 'Deadbeats')
y_2000 = y
```

```python
compare_models(x_2000, y_2000, regressors, 5)
```

* The best selection is with 2000 words,it has the same and best results than using 2813 words.
* Logisitc Regresion is our best model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_2000, y_2000, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

client_model = LogisticRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_pred_test = client_model.predict(X_test)
confusion_matrix(y_test, y_pred_test)
```

Out of 443 observations, 359 (81.03 %) are correclty classified 

```python
from sklearn import metrics

y_pred_proba = client_model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.plot(fpr,tpr)
```
