```python
# your code here
import numpy as np # Numpy
import pandas as pd # pandas
import requests # requests
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
```

```python
url = 'https://clientsfromhell.net/'
html = requests.get(url).content
soup = BeautifulSoup(html, "html.parser")
```

```python
list_items = soup.find_all('a',{'class':'ml-4 text-black hover:text-red'})
list_items
```

```python
urls = [href['href'] for href in list_items]
urls
```

```python
categories = [list_items[i].getText() for i in range(len(list_items))]
categories
```

```python
df = pd.DataFrame({
    'url': urls,
    'categories': categories
})
# sorting by url 
df.sort_values("url", inplace = True) 
  
# dropping ALL duplicte values 
df.drop_duplicates(subset ="url", inplace = True) 

df.head(20)
```

```python
df['url'] = 'https://clientsfromhell.net' + df['url'] 
df.head(10)
```

```python
df1 = df.drop(df[~df['url'].str.contains('tag')].index)
df1.head(20)
```

```python
url = 'https://clientsfromhell.net/tag/Clear-as-coal/'
def number_of_pages(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html, "html.parser")
    list_items = soup.find_all('a',{'class':'page-numbers'})
    list = [list_items[i].getText() for i in range(len(list_items))]
    return int(list[len(list)-2])

number_of_pages(url)
```

```python
df1['number_of_pages'] = df1.apply(lambda x: number_of_pages(x['url']),axis=1)
df1.head(20)
```

```python
def content(url, pages, category):
    lst = []
    for page_number in range(pages + 1):
        dynamic_url = f"{url}/page/{page_number}"
        html = requests.get(dynamic_url).content
        soup = BeautifulSoup(html, "html.parser")
        df = pd.DataFrame([str(strong_tag.text), str(strong_tag.next_sibling), category] for strong_tag in soup.find_all('strong'))
        lst.append(df)
    return pd.concat(lst).reset_index()
```

```python
list_test = []

for category in list_categories:
    url = df1[df1['categories'] == category]['url'].values[0]
    pages = df1[df1['categories'] == category]['number_of_pages'].values[0]
    list_test.append(content(url, pages, category))

df3 = pd.concat(list_test).reset_index()
```
